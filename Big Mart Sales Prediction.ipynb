### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_train['item_fat_content'] = le.fit_transform(df_train['item_fat_content'])
df_train['item_type']= le.fit_transform(df_train['item_type'])
df_train['outlet_size']= le.fit_transform(df_train['outlet_size'])
df_train['outlet_location_type']= le.fit_transform(df_train['outlet_location_type'])
df_train['outlet_type']= le.fit_transform(df_train['outlet_type'])
```

```python
df_train
```

### Splitting our data into train and test

```python
X = df_train.drop(columns='item_outlet_sales', axis=1)
```

```python
Y = df_train['item_outlet_sales']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=101, test_size=0.2)
```

```python
X_train.shape
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
X_train_std = std.fit_transform(X_train)
```

```python
X_train_std 
```

```python
X_test_std = std.transform(X_test)
```

```python
X_test_std
```

```python
Y_train
```

```python
Y_test
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
```

```python
linear.fit(X_train_std, Y_train)
```

```python
Y_test_prediction = linear.predict(X_test_std)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(Y_test,Y_test_prediction))
print(mean_absolute_error(Y_test,Y_test_prediction))
print(np.sqrt(mean_squared_error(Y_test,Y_test_prediction)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, Y_train)
```

```python
Y_test_pred = rf.predict(X_test)
```

```python
print(r2_score(Y_test,Y_test_pred))
print(mean_absolute_error(Y_test,Y_test_pred))
print(np.sqrt(mean_squared_error(Y_test,Y_test_pred)))
```
